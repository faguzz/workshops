```python
import numpy as np
from sklearn.datasets import load_sample_image
```

```python
china = load_sample_image("china.jpg")
```

```python
china.shape
```

```python
china = china.astype(np.float) / 255
```

```python
china.shape
```

```python
w, h, d = china.shape
```

```python
img = china.reshape( (w * h), d ).copy()

img_colors = img.copy()
```

```python
np.random.seed(seed=0)
np.random.shuffle(img_colors)
colors = img_colors[:1000]
```

```python
colors
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=8, random_state=0)
```

```python
km = km.fit(colors)
```

```python
km.cluster_centers_
```

```python
label = km.predict(img)
```

```python
for i in range(len(img)):
    img[i] = km.cluster_centers_[label[i]]
```

```python
new_china = img.reshape((w, h, d))
```

```python
import matplotlib.pyplot as plt

# Show original image
plt.figure(1)
plt.imshow(china)

# Show compressed image
plt.figure(2)
plt.imshow(new_china)

plt.show()
```
